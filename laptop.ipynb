```python
min_charges = df.groupby('brand')['Price'].min().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x='brand',y='Price',data=min_charges)
plt.title('Minimum price for each brand')
plt.xlabel('brand')
plt.ylabel('price')
plt.show()
```

```python
max_charges = df.groupby('brand')['Price'].max().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x='brand',y='Price',data=max_charges)
plt.title('Maximum price for each brand')
plt.xlabel('brand')
plt.ylabel('price')
plt.show()
```

```python
avg_charges = df.groupby('brand')['Price'].mean().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x='brand',y='Price',data=avg_charges)
plt.title('Average price for each brand')
plt.xlabel('brand')
plt.ylabel('price')
plt.show()
```

```python
data = df.drop(columns=['ram_type','hdd','graphic_card_gb','warranty','msoffice','rating','Number of Ratings','Number of Reviews'])
```

```python
data.head(2)
```

```python
df2 = pd.get_dummies(data,drop_first='if_binary').astype(int)
df2
```

```python
# depenedent and independent variable
x = df2.drop('Price',axis=1)
y = df2[['Price']]
x.head(2)
```

```python
x.columns
```

```python
# training and testing split..
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
# model or algorithm training..
from sklearn.ensemble import RandomForestRegressor
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
rdr = RandomForestRegressor()

# training algorithm...
rdr.fit(x_train,y_train)
print("RandomForestRegressor successfully trained.")
print()

print("RandomForestRegressor Training score: ",rdr.score(x_train,y_train) * 100)
print("RandomForestRegressor Testing score: ",rdr.score(x_test,y_test) * 100)
```

```python
import joblib,os
```

```python
os.makedirs('models',exist_ok=True)
```

```python
joblib.dump(rdr,"./models/randomforest_model.lb")
print("successfully saved your algorithm at this location : ./models/randomforest_model")
```

```python
randomforest_prediction = rdr.predict(x_test)
```

```python
compare_df = pd.concat([x_test,y_test],axis='columns')
compare_df['randomforest_prediction'] = randomforest_prediction
compare_df.head()
```

```python
os.makedirs('prediction',exist_ok=True)
```

```python
compare_df.to_csv('./prediction/.prediction.csv',index=False)
print('saved')
```
